```python
import pandas as pd

# Load the dataset
file_path = '/Users/minhtao/Downloads/My MacBook/AI and Data Science/Forage/BCGXGenAI/BCGX_Gen_AI/Task 1. Data Extraction and Initial Analysis/Final_report.csv'
df = pd.read_csv(file_path)

# Display the first few rows of the dataframe to understand its structure
df.head()
```

# Columns:
Year: Fiscal year of the data.
Total Revenue: The total revenue for the company in that fiscal year.
Net Income: The net income for the company in that fiscal year.
Total Assets: The total assets for the company in that fiscal year.
Total Liabilities: The total liabilities for the company in that fiscal year.
Cash Flow from Operating Activities: The cash flow from operating activities for the company in that fiscal year.
Company: The name of the company (Microsoft, Apple, or Tesla).

```python
# Calculate year-over-year growth for each financial metric
df['Revenue Growth (%)'] = df.groupby('Company')['Total Revenue'].pct_change() * 100
df['Net Income Growth (%)'] = df.groupby('Company')['Net Income'].pct_change() * 100
df['Assets Growth (%)'] = df.groupby('Company')['Total Assets'].pct_change() * 100
df['Liabilities Growth (%)'] = df.groupby('Company')['Total Liabilities'].pct_change() * 100
df['Cash Flow Growth (%)'] = df.groupby('Company')['Cash Flow from Operating Activities'].pct_change() * 100

# Display the updated dataframe
df
```

```python
# Imputing Null/NaN values that result from pct_change calculations with 0
df.fillna(0, inplace=True)
df
```

```python
df.to_csv('Financial_data_with_growth_rates.csv')
```

# Summary of Calculated Growth Rates:
Revenue Growth (%): The percentage change in total revenue from one fiscal year to the next.
Net Income Growth (%): The percentage change in net income from one fiscal year to the next.
Assets Growth (%): The percentage change in total assets from one fiscal year to the next.
Liabilities Growth (%): The percentage change in total liabilities from one fiscal year to the next.
Cash Flow Growth (%): The percentage change in cash flow from operating activities from one fiscal year to the next.

```python
# Compute the average growth for each financial metric for each company
avg_growth = df.groupby('Company').mean().drop(columns=['Year', 'Total Revenue', 'Net Income', 'Total Assets', 'Total Liabilities', 'Cash Flow from Operating Activities'])

# Display the average growth for each company
avg_growth
```

```python
avg_growth.to_csv('Average_growth_rates_by_company.csv')
```
